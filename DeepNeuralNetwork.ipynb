```python
number_of_inputs = 8
train_split = 0.8
val_split = 0.1
test_split = 0.1
iodataset = dataFrame.values
np.random.shuffle(iodataset)
```

```python
indices_or_sections = [int(train_split * len(dataFrame)), int((1 - test_split) * len(dataFrame))]
train_set, val_set, test_set = np.split(iodataset, indices_or_sections)
np.set_printoptions(suppress=True)
print("Train set size: ", len(train_set))
print("Validation set size: ", len(val_set))
print("Test set size: ", len(test_set))
x_train = train_set[:,0:number_of_inputs]
y_train = train_set[:,number_of_inputs]
x_val = val_set[:,0:number_of_inputs]
y_val = val_set[:,number_of_inputs]
x_test = test_set[:,0:number_of_inputs]
y_test = test_set[:,number_of_inputs]
```

```python
import numpy as np

class my_NN(object):
    def __init__(self):
        self.input = 8  # feature numbers
        self.output = 1  # class number
        self.hidden_units = 4

        # initialize matrix of weights
        np.random.seed(1)
        # weight1: input = hidden layer
        self.w1 = np.random.randn(self.input, self.hidden_units)
        # weight2: hidden layer = output
        self.w2 = np.random.randn(self.hidden_units, self.output)

    def _normalize(self, x):
        # Min-Max normalization to [0, 1] range
        return (x - np.min(x)) / (np.max(x) - np.min(x))

    # forward_propagation
    def _forward_propagation(self, x_train):
        # Normalize
        normalized_x_train = self._normalize(x_train)

        self.z2 = np.dot(self.w1.T, normalized_x_train.T)
        self.a2 = self._sigmoid(self.z2)
        self.z3 = np.dot(self.w2.T, self.a2)
        self.a3 = self._sigmoid(self.z3)
        return self.a3

    def _sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def _loss(self, predict, y_train):
        m = y_train.shape[0]
        logprobs = np.multiply(np.log(predict), y_train) + np.multiply((1 - y_train), np.log(1 - predict))
        loss = -np.sum(logprobs) / m
        return loss

    # backpropagation
    def _backward_propagation(self, x_train, y_train):
        predict = self._forward_propagation(x_train)
        m = x_train.shape[0]
        delta3 = predict - y_train
        dz3 = np.multiply(delta3, self._sigmoid_prime(self.z3))
        self.dw2 = (1 / m) * np.sum(np.multiply(self.a2, dz3), axis=1).reshape(self.w2.shape)

        delta2 = delta3 * self.w2 * self._sigmoid_prime(self.z2)
        self.dw1 = (1 / m) * np.dot(x_train.T, delta2.T)

    def _sigmoid_prime(self, z):
        return self._sigmoid(z) * (1 - self._sigmoid(z))

    def _update(self, learning_rate=0.01):
        self.w1 = self.w1 - learning_rate * self.dw1
        self.w2 = self.w2 - learning_rate * self.dw2

    def train(self, x_train, y_train, iteration=10):
        for i in range(iteration):
            y_hat = self._forward_propagation(x_train)
            loss = self._loss(y_hat, y_train)
            self._backward_propagation(x_train, y_train)
            self._update()
            if i % 10 == 0:
                print("loss: ", loss)

    def predict_proba(self, x_train):
        y_hat = self._forward_propagation(x_train)
        return y_hat.T

    def predict(self, x_train):
        y_hat_proba = self.predict_proba(x_train)
        return [1 if i[0] >= 0.5 else 0 for i in y_hat_proba]

    def score(self, predict, y_train):
        cnt = np.sum(predict == y_train)
        return (cnt / len(y_train)) * 100


if __name__ == '__main__':

    clr = my_NN()
    clr.train(x_train, y_train)
    pre_y_proba = clr.predict_proba(x_test)
    pre_y = clr.predict(x_test)
    # Accuracy
    score = clr.score(pre_y, y_test)

    print('Predicted Probabilities: ', pre_y_proba)
    print('Predicted Classes: ', pre_y)
    print('True Classes: ', y_test)
    print('Accuracy Score: ', score)
```
